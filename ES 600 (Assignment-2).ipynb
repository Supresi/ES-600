# Sphereman Corellation between parameter of CT (Qu -1)

```python
from scipy.stats import spearmanr

ct_spearman_correlation_matrix1 = pd.DataFrame(index=ct.columns, columns=ct.columns)

for ct_column in l_ct:
    for ct_column_1 in l_ct:
#         print(ct_column," ",gw_column)
#          aa = ct[ct_column].corr(gw[gw_column])
         ct_spearman_correlation_matrix1.loc[ct_column, ct_column_1],p_val = spearmanr(ct[ct_column],ct[ct_column_1])



print("Correlation matrix between ct :")

# print(correlation_matrix1['PH']['PH'])
# print('\n')
# print(ct_spearman_correlation_matrix1)

print(type(ct_spearman_correlation_matrix1))
```
